```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BTC = pd.read_csv('BTC_reshaped.csv').drop(['Unnamed: 0'], axis=1)
ETH = pd.read_csv('ETH_reshaped.csv').drop(['Unnamed: 0'], axis=1)
XRP = pd.read_csv('XRP_reshaped.csv').drop(['Unnamed: 0'], axis=1)
DOGE = pd.read_csv('DOGE.csv').drop(['Unnamed: 0'], axis=1)
VET = pd.read_csv('VET.csv').drop(['Unnamed: 0'], axis=1)
TUSD = pd.read_csv('TUSD.csv').drop(['Unnamed: 0'], axis=1)
XVG = pd.read_csv('XVG.csv').drop(['Unnamed: 0'], axis=1)
AE = pd.read_csv('AE.csv').drop(['Unnamed: 0'], axis=1)
STRAT = pd.read_csv('STRAT.csv').drop(['Unnamed: 0'], axis=1)
LRC = pd.read_csv('LRC.csv').drop(['Unnamed: 0'], axis=1)
AOA = pd.read_csv('AOA.csv').drop(['Unnamed: 0'], axis=1)
QNT = pd.read_csv('QNT.csv').drop(['Unnamed: 0'], axis=1)
WAN = pd.read_csv('WAN_reshaped.csv').drop(['Unnamed: 0'], axis=1).drop(['Unnamed: 0.1'], axis=1)
ELA = pd.read_csv('ELA_reshaped.csv').drop(['Unnamed: 0'], axis=1).drop(['Unnamed: 0.1'], axis=1)
KCS = pd.read_csv('KCS_reshaped.csv').drop(['Unnamed: 0'], axis=1).drop(['Unnamed: 0.1'], axis=1)
```

```python
QNT.head()
```

```python
BTC = BTC.rename(columns={"Close Percent Change": "CPC1"})
ETH = ETH.rename(columns={"Close Percent Change": "CPC2"})
XRP = XRP.rename(columns={"Close Percent Change": "CPC3"})
DOGE = DOGE.rename(columns={"Close Percent Change": "CPC4"})
VET = VET.rename(columns={"Close Percent Change": "CPC5"})
TUSD = TUSD.rename(columns={"Close Percent Change": "CPC6"})
XVG = XVG.rename(columns={"Close Percent Change": "CPC7"})
AE = AE.rename(columns={"Close Percent Change": "CPC8"})
STRAT = STRAT.rename(columns={"Close Percent Change": "CPC9"})
LRC = LRC.rename(columns={"Close Percent Change": "CPC10"})
AOA = AOA.rename(columns={"Close Percent Change": "CPC11"})
QNT = QNT.rename(columns={"Close Percent Change": "CPC12"})
WAN = WAN.rename(columns={"Close Percent Change": "CPC13"})
ELA = ELA.rename(columns={"Close Percent Change": "CPC14"})
KCS = KCS.rename(columns={"Close Percent Change": "CPC15"})
```

```python
pct_change = pd.concat([BTC.loc[:, ['Date']], BTC.loc[:, ['CPC1']],
                        ETH.loc[:, ['CPC2']], XRP.loc[:, ['CPC3']],
                        DOGE.loc[:, ['CPC4']], VET.loc[:, ['CPC5']],
                        TUSD.loc[:, ['CPC6']], XVG.loc[:, ['CPC7']],
                        AE.loc[:, ['CPC8']], STRAT.loc[:, ['CPC9']],
                        LRC.loc[:, ['CPC10']], AOA.loc[:, ['CPC11']],
                        QNT.loc[:, ['CPC12']], WAN.loc[:, ['CPC13']],
                        ELA.loc[:, ['CPC14']], KCS.loc[:, ['CPC15']]], axis=1, join_axes=[BTC.index])                           
                           
                           
```

```python
pct_change = pct_change.rename(columns={ pct_change.columns[1]: "BTC Pct Change", 
                                        pct_change.columns[2]: "ETH Pct Change",
                                       pct_change.columns[3]: "XRP Pct Change",
                                        pct_change.columns[4]: "DOGE Pct Change",
                                        pct_change.columns[5]: "VET Pct Change",
                                        pct_change.columns[6]: "TUSD Pct Change",
                                        pct_change.columns[7]: "XVG Pct Change",
                                        pct_change.columns[8]: "AE Pct Change",
                                        pct_change.columns[9]: "STRAT Pct Change",
                                        pct_change.columns[10]: "LRC Pct Change",
                                        pct_change.columns[11]: "AOA Pct Change",
                                        pct_change.columns[12]: "QNT Pct Change",
                                        pct_change.columns[13]: "WAN Pct Change",
                                       pct_change.columns[14]: "ELA Pct Change",
                                       pct_change.columns[15]: "KCS Pct Change"})
```

## Making a Copy of the data frame is important

```python
pct_change_tf = pct_change.copy()
```

```python
pct_change['BTC Pct Change'][0]
```

```python
pct_change_tf['BTC Pct Change'][0]
```

```python
len(pct_change['BTC Pct Change'])
```

```python
pct_change_tf.columns
```

```python
for col in pct_change.columns:
    if col != 'Date':
        for i in list(range(0, len(pct_change[col]))):
            if pct_change[col][i] > 0:
                pct_change_tf[col][i] =True
            else: 
                pct_change_tf[col][i] =False
    
```

```python
pct_change_tf.head()
```

```python
tot_pos = {}
for col in pct_change_tf.columns:
    if col != 'Date':
        tot_pos[col] = round((sum(pct_change_tf[col])/len(pct_change_tf[col]))*100, 2)
```
